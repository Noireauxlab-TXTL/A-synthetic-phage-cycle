```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the bar plot
conditions = ["+LPS", "-LPS"]
phages = ["T7mC-S*", "T7 WT"]
values = [[0.031, 0.08], [0.54, 0.14]]  # Mean values
errors = [[0.017, 0.05], [0.21, 0.04]]  # Error values

# Bar positions
x = np.arange(len(conditions))
bar_width = 0.35

# Additional data points for individual measurements
colors=['lightcoral', 'gray']
data_points = {
    "T7mC-S*": {
        "+LPS": [0.00251046*106026785.7/21875000, 0.009386733*106026785.7/21875000, 0.007125891*106026785.7/21875000],
        "-LPS": [0.028051683*106026785.7/21875000, 0.01233616*106026785.7/21875000, 0.007026627*106026785.7/21875000]
    },
    "T7 WT": {
        "+LPS": [0.662910176, 0.656311529, 0.292141396],
        "-LPS": [0.178794922, 0.119296153, 0.108760672]
    }
}

# Create the bar plot with individual data points
#fig, ax = plt.subplots(figsize=(8, 6))
fig, ax = plt.subplots(figsize=(5, 6))
for i, (phage, means, errs) in enumerate(zip(phages, values, errors)):
    bar_positions = x + i * bar_width
    # Plot bars
    ax.bar(bar_positions, means, bar_width, label=phage, yerr=errs, capsize=5, edgecolor='black', color=colors)
    # Plot individual data points
    for j, condition in enumerate(conditions):
        points = data_points[phage][condition]
        jitter = np.linspace(-0.1, 0.1, len(points))  # Distribute points horizontally on the bar
        ax.scatter(bar_positions[j] + jitter, points, color='black', zorder=5)

# Customize the plot
ax.set_xticks(x + bar_width / 2)
ax.set_xticklabels(conditions)
ax.set_ylabel("PFU per liposome at 0.05 nM")
ax.set_title("PFU per liposome by TXTL condition and phage type")
ax.legend(title="Phage type")
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
#plt.savefig("Fig rebinding.svg", format='svg')
plt.show()
```
